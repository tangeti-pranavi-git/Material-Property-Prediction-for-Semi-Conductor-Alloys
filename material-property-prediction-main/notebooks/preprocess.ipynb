```python
import pandas as pd

csv_path = r"D:\MDS\MDS CBP\data\MDS CBP\combined_file.csv"
df = pd.read_csv(csv_path)
print("Initial shape:", df.shape)
```

```python

# 1. Check Missing Values
print("Missing per column:\n", df.isnull().sum())
```

```python

# 2. Drop rows missing any key target property
target_columns = ["band_gap", "density", "formation_energy_per_atom", "efermi"]  # Add/adjust columns
df_clean = df.dropna(subset=target_columns)
print("Shape after dropping missing targets:", df_clean.shape)
```

```python

# 3. Only drop truly unnecessary columns (do NOT drop formula_pretty!)
columns_to_keep = ["material_id", "formula_pretty", "band_gap", "density", "formation_energy_per_atom", "efermi"]
columns_to_keep = [col for col in columns_to_keep if col in df_clean.columns]
df_clean = df_clean[columns_to_keep + [c for c in df_clean.columns if c not in columns_to_keep]]  # Keep all columns for now
```

```python

# 4. Remove Duplicates
df_clean = df_clean.drop_duplicates()
print("Shape after dropping duplicates:", df_clean.shape)
```

```python

# 5. Filter out unphysical values
df_clean = df_clean[df_clean["density"] > 0]
df_clean = df_clean[df_clean["band_gap"] >= 0]
```

```python

# 6. Reset Index
df_clean = df_clean.reset_index(drop=True)
print("Final cleaned shape:", df_clean.shape)
print("Columns:", df_clean.columns.to_list())

```

```python

# 7. Save Cleaned Data
cleaned_path = r"D:\MDS\MDS CBP\data\MDS CBP\cleaned_combined_file.csv"
df_clean.to_csv(cleaned_path, index=False)
print(f"Cleaned data saved to: {cleaned_path}")
```
